# Q5 - Find the top 3 Airlines which had the largest mean departure delay in the month of January (2 points)
Month of January has the id of 1. Consider only flights with the positive departure delay. 

```python
# Add your code here!
# create new df with only January flights with positive departure delay
q5=df[(df["MONTH"]==1)&(df["DEPARTURE_DELAY"]>0)]
q5["DEPARTURE_DELAY"].mean()
q5
```

```python
q5_group = q5.groupby("AIRLINE")["DEPARTURE_DELAY"].mean()
q5_group.sort_values(ascending=False, inplace=True)
q5_group

# The top 3 airlines with the highest average departure delay in January are: F9, OO, and MQ
```

# Q7 - Which origin airport has the most cancellation of departure flights? (2 points)


```python
# Add your code here! 
# group dataframe by the sum of cancelled flights for each Origin Airport
q7=df.groupby("ORIGIN_AIRPORT")["CANCELLED"].sum()
q7.sort_values(ascending=False, inplace=True)
q7

# The Origin Airport with the highest number of cancellations is ORD
```

## Q8 - Create a histogram plot that shows the data distribution of the arrival delays consider only positive arrival delays and smaller than 200 minutes. Use 10 minutes bin/bucket size. (2 points)

```python
# Add your code here! 
# create a new dataframe with arrival delay between 0 and 200 minutes
q8=df[df["ARRIVAL_DELAY"]>0]
q8=q8[q8["ARRIVAL_DELAY"].between(0, 200)]
q8
```

```python
# create histogram of arrival delays between 0 and 200 minutes with bins of 10 minutes
plt.hist(q8["ARRIVAL_DELAY"], bins=20)
plt.xlabel("Arrival Delay (minutes)")
plt.ylabel("Frequency")
plt.title("Arrival Delay Distribution")
plt.xticks(ticks=np.arange(0, 200, step=10), 
           labels=np.arange(0, 200, step=10),
           rotation=90)
plt.show()
```
